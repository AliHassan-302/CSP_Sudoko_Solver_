```python
# print board nicely
def print_board(board):
    for row in board:
        print(" ".join(str(num) for num in row))

# check if number can be placed in (row, col, box)
def is_safe(board, row, col, num):
    for x in range(9):
        if board[row][x] == num:
            return False
    for x in range(9):
        if board[x][col] == num:
            return False
    sr, sc = row - row % 3, col - col % 3
    for i in range(3):
        for j in range(3):
            if board[sr + i][sc + j] == num:
                return False
    return True

# find empty cell (0 means empty)
def find_empty(board):
    for i in range(9):
        for j in range(9):
            if board[i][j] == 0:
                return i, j
    return None

# counters for report
backtrack_calls = 0
failures = 0

# backtracking search (DFS)
def solve(board):
    global backtrack_calls, failures
    backtrack_calls += 1

    empty = find_empty(board)
    if not empty:
        return True

    row, col = empty
    for num in range(1, 10):
        if is_safe(board, row, col, num):
            board[row][col] = num
            if solve(board):
                return True
            board[row][col] = 0

    failures += 1
    return False

# basic AC3 placeholder
def AC3(board):
    return True

# run solver with hardcoded boards (easy/medium)
def run_sudoku(level):
    global backtrack_calls, failures
    backtrack_calls = 0
    failures = 0

    if level == "easy":
        board = [
            [0,0,4,0,3,0,0,5,0],
            [6,0,9,4,0,0,0,0,0],
            [0,0,5,1,0,0,4,8,9],
            [0,0,0,0,6,0,9,3,0],
            [3,0,0,8,0,7,0,0,2],
            [0,2,6,0,4,0,0,0,0],
            [4,5,3,0,0,9,6,0,0],
            [0,0,0,0,0,4,7,0,5],
            [0,9,0,0,5,0,2,0,0]
        ]
    elif level == "medium":
        board = [
            [0,0,0,2,6,0,7,0,1],
            [6,8,0,0,7,0,0,9,0],
            [1,9,0,0,0,4,5,0,0],
            [8,2,0,1,0,0,0,4,0],
            [0,0,4,6,0,2,9,0,0],
            [0,5,0,0,0,3,0,2,8],
            [0,0,9,3,0,0,0,7,4],
            [0,4,0,0,5,0,0,3,6],
            [7,0,3,0,1,8,0,0,0]
        ]
    else:
        print("wrong level")
        return

    print("\nOriginal Board:\n")
    print_board(board)

    AC3(board)

    if solve(board):
        print("\nSolved Board:\n")
        print_board(board)
    else:
        print("No solution found")

    print("\nBacktrack calls:", backtrack_calls)
    print("Failures:", failures)
    
run_sudoku("easy")
```
